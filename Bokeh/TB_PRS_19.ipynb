# Various Tools 

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 4*np.pi, 200)
y = np.sin(x)
```

```python
fig = figure(
    title="Bokeh Tools Example",
    tools="pan,box_zoom,wheel_zoom,reset,save,hover,crosshair, lasso_select",
    toolbar_location="below", toolbar_sticky=False, width=1000, height=400
    )
fig.line(x, y, line_width=2)
fig.circle(x, y, size=5, color="red", alpha=0.5)
show(fig)
```

# Text Properties, Fill Properties and Styling Visual Attributes 

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

fig = figure(width=800, height=400, title="Text Properties, Fill Properties and Styling Visual Attributes")

fig.line(x, y, legend_label="Sine x")
fig.circle(x, y, size=5, legend_label="Sine x")

fig.line(x, 2*y, legend_label="2 * Sine x", color="orange",line_width=2, line_dash="dashed")

fig.square(x, 3*y, legend_label="3 * Sine x", line_color="green", alpha=0.5, fill_color=None)
fig.line(x, 3*y, legend_label="3 * Sine x", line_color="green", alpha=0.5, line_width=2)

fig.legend.location = "bottom_left"

show(fig)
```

```python
from bokeh.models import Legend
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

fig = figure(width=800, height=400, title="Text Properties, Fill Properties and Styling Visual Attributes", toolbar_location="above")

r0=fig.line(x, y)
r1=fig.circle(x, y, size=5)

r2=fig.line(x, 2*y, color="orange",line_width=2, line_dash="dashed")

r3=fig.square(x, 3*y, line_color="green", alpha=0.5, fill_color=None)
r4=fig.line(x, 3*y, line_color="green", alpha=0.5, line_width=2)

legend = Legend(items=[
    ("Sine x", [r0]),
    ("Sine x (circle)", [r1]),
    ("2 * Sine x", [r2]),
    ("3 * Sine x (square)", [r3]),
    ("3 * Sine x", [r4]),
])
fig.add_layout(legend, "right")

show(fig)
```
