````python
import numpy as np
import matplotlib.pyplot as plt     
import pandas as pd
````

# Contour
"contour" refers to a contour plot, which is a two-dimensional graphical representation of a three-dimensional surface. It visualizes a function of two variables, \(Z=f(X,Y)\), by drawing lines (or filling areas) that connect points of equal value of \(Z\)

````python
xlist=np.linspace(-10, 10, 100)
ylist=np.linspace(-10, 10, 100)

X, Y = np.meshgrid(xlist, ylist) # Create a meshgrid
Z = np.sqrt(X**2 + Y**2)

plt.figure(figsize=(10, 4))
plt.contourf(X, Y, Z) # filled meshgrid
# plt.contour(X, Y, Z) # not filled meshgrid
plt.colorbar()
plt.title('Contour Plot of Z')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')    
plt.show()
````

## What is Annotation in Matplotlib?

Annotation in Matplotlib refers to adding explanatory text, arrows, or other markers to a plot to highlight specific data points or features. This helps make plots more informative and easier to interpret.

- **Text Annotation:** Adds descriptive text at a specific location.
- **Arrow Annotation:** Points to a feature using an arrow.

**Example:**
```python
plt.annotate('Important Point', xy=(x_value, y_value), xytext=(x_text, y_text), arrowprops=dict(facecolor='red'))
```

- `xy`: Coordinates of the point to annotate.
- `xytext`: Location of the annotation text.
- `arrowprops`: Properties of the arrow (style, color, etc.).

Annotations are useful for emphasizing trends, outliers, or key observations in your visualizations.

````python
x=np.array([1,2,3,4,5])
y=np.array([20,33,11,25,41])

plt.figure(figsize=(10, 4))
plt.plot(x, y, marker='v', linestyle='--', color='g')
plt.annotate('Mar Down', xy=(4, 25), xytext=(4.5, 20), arrowprops=dict(facecolor='red', shrink=0.05))
plt.title('Line Plot with Annotation')
plt.show()
````

````python
x=np.array([1,2,3,4,5])
y=np.array([20,33,11,25,41])

plt.figure(figsize=(10, 4))
plt.scatter(x, y, marker='v', linestyle='--', color='g')
plt.annotate('Mar Down', xy=(4, 25), xytext=(4.5, 20), arrowprops=dict(facecolor='red', shrink=0.05))
plt.title('Scatter Plot with Annotation')
plt.show()
````

# Advanced Annotation

### Box Styles in Matplotlib Annotation ###

Matplotlib provides several box styles for annotations using the `boxstyle` parameter in `bbox`. Here are some common types:

- `'square'` : Square box
- `'round'` : Rounded corners
- `'round4'` : More rounded corners
- `'darrow'` : Double arrow
- `'larrow'` : Left arrow
- `'rarrow'` : Right arrow
- `'sawtooth'` : Sawtooth edge
- `'roundtooth'` : Rounded sawtooth edge
- `'circle'` : Circular box
- `'lpad'` : Padding on the left
- `'rpad'` : Padding on the right

**Example usage:**
```python
bbox_props = dict(boxstyle='round,pad=0.5', fc='yellow', ec='black')
plt.text(2, 30, 'Example', bbox=bbox_props)
```

You can customize the appearance by combining box styles with padding and colors. For more details, see [Matplotlib documentation on box styles](https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_bbox.html).

````python
x=np.array([1,2,3,4,5])
y=np.array([20,33,11,25,41])

plt.figure(figsize=(10, 4))
plt.plot(x, y, marker='v', linestyle='--', color='g')
bbox_props= dict(boxstyle='round,pad=0.3', facecolor='cyan')
plt.text(4, 20, 'Direction', ha='center', va='center', size=10, bbox=bbox_props, rotation=45)
plt.xlabel('X Axis')
plt.show()
````

## Connection Styles in Matplotlib Annotation

**Connection style** in Matplotlib annotation controls the appearance of the arrow connecting the annotation text to the point of interest. It is set using the `connectionstyle` parameter inside `arrowprops` in `plt.annotate()`.

### Common Connection Styles

- `'arc3'` : Simple curved arc (default)
- `'angle'` : Straight lines with a sharp angle
- `'angle3'` : Three-segment angle
- `'bar'` : Horizontal bar at the base
- `'arc'` : Circular arc
- `'fraction'` : fraction of bar

**Example Usage:**
```python
plt.annotate(
    'Point',
    xy=(x_value, y_value),
    xytext=(x_text, y_text),
    arrowprops=dict(
        arrowstyle='->',
        connectionstyle='arc3,rad=0.2'
    )
)
```

### Customizing Connection Styles

You can further customize styles with parameters:
- `rad` (for `'arc3'`): Curvature radius
- `angleA`, `angleB` (for `'angle'`): Angles at start and end
- `armA`, `armB` (for `'angle3'`): Lengths of arms

**More Info:**  
See [Matplotlib documentation on connection styles](https://matplotlib.org/stable/gallery/text_labels_and_annotations/arrow_demo.html) for details and visual examples.

````python
x=np.array([1,2,3,4,5])
y=np.array([20,33,11,25,41])

plt.figure(figsize=(10, 4))
plt.plot(x, y, marker='v', linestyle='--', color='g')
plt.annotate('Mark Down', xy=(4, 25), xytext=(4.5, 20), arrowprops=dict(fc='r',shrink=0.05, connectionstyle='arc3,rad=0.5'))
plt.title('Scatter Plot with Annotation')
plt.show()
````

## Arrow Styles in Matplotlib Annotation

The `arrowstyle` parameter in Matplotlib controls the appearance of arrows in annotations. It is used inside the `arrowprops` dictionary in `plt.annotate()`.

### Common Arrow Styles

- `'-'` : Simple line
- `'->'` : Arrow pointing right
- `'-['` : Bar at the end
- `'<-'` : Arrow pointing left
- `'<->'` : Double-headed arrow
- `'fancy'` : Fancy arrow
- `'simple'` : Simple arrow
- `'wedge'` : Wedge-shaped arrow

**Example Usage:**
```python
plt.annotate(
    'Point',
    xy=(x_value, y_value),
    xytext=(x_text, y_text),
    arrowprops=dict(arrowstyle='->', color='red')
)
```

You can further customize arrows with parameters like width, head_length, and head_width.  
See [Matplotlib documentation on arrow styles](https://matplotlib.org/stable/gallery/text_labels_and_annotations/arrow_demo.html) for more details and visual examples.

````python
x=np.array([1,2,3,4,5])
y=np.array([20,33,11,25,41])

plt.figure(figsize=(10, 4))
plt.plot(x, y, marker='v', linestyle='--', color='g')
plt.annotate('Mar Down', xy=(4, 25), xytext=(4.5, 20), arrowprops=dict(arrowstyle='wedge', facecolor='red', connectionstyle='arc3,rad=0.5'))
plt.title('Scatter Plot with Annotation')
plt.show()
````
