```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# Using ax method
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.show() 
```

```python
# Using plt method
plt.axes(projection='3d')
plt.show()
```

```python
#  Using plt method
plt.rcParams['legend.fontsize'] = 10 # Setting legend font size

plt.axes(projection='3d')

x = [1, 2, 3, 4, 5]
y = [20, 33, 11, 25, 41]
z = [0, 0, 0, 0, 0]   # keep z axis flat

plt.plot(x, y, z, label='3D Line', color='g')
plt.legend()

plt.show()
```
