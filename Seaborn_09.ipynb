### ci In Seaborn
In seaborn, `ci` stands for "confidence interval". It is a parameter used in plotting functions (such as `sns.lineplot`, `sns.barplot`, etc.) to indicate the size of the confidence interval to draw around estimated values. By default, seaborn shows a 95% confidence interval (`ci=95`). You can set `ci=None` to hide the confidence interval, or specify another value (e.g., `ci=68` for a 68% interval).

**Example usage:**
```python
sns.lineplot(x='total_bill', y='tip', data=df, ci=95)
```

**Common values for `ci`:**
- `ci=95`: 95% confidence interval (default)
- `ci=None`: No confidence interval  
- `ci="sd"`: Show standard deviation instead of confidence interval
- `ci=68`: 68% confidence interval

The confidence interval provides a visual indication of the uncertainty around the estimated values.

````python
sns.relplot(x='day', y='tip', data=df, kind='line')
plt.show()
````

````python
sns.relplot(x='day', y='tip', data=df, kind='line', ci='sd')
sns.relplot(x='day', y='tip', data=df, kind='line', ci=None)
sns.relplot(x='day', y='tip', data=df, kind='line', ci=68)
plt.show()
````
