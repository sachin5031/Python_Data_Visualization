```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
height = [105, 110, 115, 120, 125, 130, 135, 140, 145, 150]
weight = [20, 22, 24, 26, 28, 30, 32, 34, 36, 38]   

sns.scatterplot(x=height, y=weight)
```

```python
gender= ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male','Male', 'Male']


sns.countplot(x=gender, hue=gender)
plt.show()
```

```python
df=sb.load_dataset("tips")

df
```

```python
df.columns
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='smoker')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='smoker', hue_order=['No', 'Yes'])
plt.show()
```

```python
hue_color = {'Yes': 'black', 'No': 'red'}
sns.scatterplot(x='total_bill', y='tip', data=df, hue='smoker', hue_order=['Yes', 'No'], palette=hue_color)
plt.show()
```
