# Seaborn
    Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn makes it easy to visualize data distributions, relationships, and categorical data, and it integrates well with pandas DataFrames for quick plotting and analysis.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sb
````

````python
height = [105, 110, 115, 120, 125, 130, 135, 140, 145, 150]
weight = [20, 22, 24, 26, 28, 30, 32, 34, 36, 38]   

sns.scatterplot(x=height, y=weight)
````

````python
gender= ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male','Male', 'Male']


sns.countplot(x=gender, hue=gender)
plt.show()
````

### relplot in Seaborn

`relplot` is a figure-level function in Seaborn used for visualizing relationships between variables. It can create scatter plots (`kind='scatter'`) and line plots (`kind='line'`). `relplot` combines the flexibility of FacetGrid with the convenience of high-level plotting, allowing for easy creation of multi-plot grids based on categorical variables.

**Example:**
```python
sns.relplot(x='total_bill', y='tip', data=df, kind='scatter', hue='smoker')
```

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='scatter', hue='smoker')
plt.show()
````

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='line', hue='smoker')
plt.show()
````

### Subplot in Seaborn
You can use the `row` and `col` parameters in `relplot` to create subplots based on categorical variables. For example, `row='sex'` and `col='day'` will create a grid of scatter plots, with rows for each sex and columns for each day. This helps visualize how relationships differ across categories.

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='scatter', row='smoker')
plt.show()
````

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='scatter', col='smoker')
plt.show()
````

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='line', col='smoker')
plt.show()
````

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='line', col='sex')
plt.show()
````

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='line', col='day')
plt.show()
````

````python
sns.relplot(x='total_bill', y='tip', data=df, kind='line', col='time')
plt.show()
````
