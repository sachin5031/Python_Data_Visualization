# Twin Axis Plot

    A twin axis plot (also called a dual-axis plot) is a type of chart that displays two different y-axes on the same graph, sharing a common x-axis. This allows you to visualize two datasets with different scales or units together for easy comparison.

```python
# 8,9/08/2025
import matplotlib.pyplot as plt
import pandas as pd  
import numpy as np
```

```python
df = pd.read_csv('data.csv',parse_dates=True)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.set_index('date')
```

```python
df.head()
```

```python
df.plot(figsize=(10,5),title="Time Series Plot",fontsize=15) # It will not show the full date 
```

```python
fig, ax = plt.subplots()

# First axis (Price)
ax.plot(df.index, df['price'], color='b', label='Price')
ax.set_xlabel('Date')
ax.set_ylabel('Price', color='b')
ax.set_ylim(200, 2000)
ax.tick_params(axis='y', labelcolor='b')  # changes text color

# Second axis (Sales)
ax2 = ax.twinx()
ax2.plot(df.index, df['sales'], color='r', label='Sales')
ax2.set_ylabel('Sales', color='r')
ax2.set_ylim(1000, 10000)
ax2.tick_params(axis='y', labelcolor='r')  # changes text color

# Legends
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# ax.legend()
# ax2.legend()

plt.show()
```

```python
plt.plot(df['price'], color='b', label='Price')
plt.plot(df['sales'], color='r', label='Sales')
plt.title('Twin Axis Plot')
plt.show()
```
