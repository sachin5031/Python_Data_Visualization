# Box Plot

## whis
The `whis` parameter in Seaborn's box plot controls the length of the whiskers. Whiskers extend from the box to the data points that are within a specified distance from the quartiles, typically 1.5 times the interquartile range (IQR). You can set `whis` to a float, list, or percentile to customize how far the whiskers reach.

## sym
The `sym` parameter specifies the marker style for outliers in the box plot. By default, outliers are shown as individual points. You can change the symbol or hide outliers by setting `sym=''`.


````python
sns.catplot(x='day', y='total_bill', data=df, kind='box', hue='day')
plt.show()
````

````python
sns.catplot(x='time', y='total_bill', data=df, kind='box', hue='time', whis=[0,100])
plt.show()
````

````python
sns.catplot(x='time', y='total_bill', data=df, kind='box', hue='time', flierprops={"marker": ""})
plt.show()
````

````python
sns.catplot(x='time', y='total_bill', data=df, kind='box', hue='time', flierprops={"marker": ""}, whis=[0,100])
plt.show()
````

````python
sns.catplot(x='time', y='total_bill', data=df, kind='box', hue='time', order=['Dinner', 'Lunch'], flierprops={"marker": ""}, whis=[0,100])
plt.show()
````

## Point Plot in Seaborn

A **point plot** in Seaborn is a visualization used to show the mean value of a quantitative variable for one or more categories, with error bars representing uncertainty (such as standard deviation or confidence interval). It is useful for comparing values across categories and visualizing trends.

- **Function:** `sns.pointplot()`
- **Typical Usage:** Compare means across groups, visualize categorical data with error bars.
- **Features:** Displays points for means, connects them with lines, and shows error bars.

**Example:**
```python
sns.pointplot(x='day', y='total_bill', data=df)
plt.show()
```

This will plot the average `total_bill` for each `day` in the `df` dataset, with error bars.
