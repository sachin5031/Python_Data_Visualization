# Bar, Box, Violin, Histogram and Scatter Plotting

```python
# 11/8/2025
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

# Bar Plotting

```python
# Bar Plots

movie_type = ['comedy', 'romance', 'drama', 'action', 'scifi']
rating = [4, 3, 5, 7, 8]

plt.figure(figsize=(10, 5))
plt.bar(movie_type, rating, color='r', width=0.3)
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.ylabel("Rating")
plt.title("Movie Ratings")
plt.show()
```

```python
#Bar Plotting using error

plt.figure(figsize=(8,4))
plt.bar(movie_type, rating, color='r', width=0.3, yerr=True)
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.xlabel("Movie Type")
plt.ylabel("Rating")
plt.title("Movie Ratings with Error Bars")
plt.show()
```

```python
# Using Statistics in error
error_mean=statistics.mean(rating)
error_std=statistics.stdev(rating)

plt.figure(figsize=(6,3))
# plt.bar(movie_type, rating, color='g', width=0.3, yerr=error_std)
plt.bar(movie_type, rating, color='g', width=0.3, yerr=error_mean)
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.xlabel("Movie Type")
plt.ylabel("Rating")
plt.title("Movie Ratings with Error Bars")
plt.show()
```

```python
# Bar chart using gender wise
men_rating = [4, 3, 5, 7, 8]
women_rating = [5, 4, 6, 8, 9]

plt.figure(figsize=(8, 4))
plt.bar(movie_type, men_rating, color='purple', width=0.4, label='Men')
plt.bar(movie_type, women_rating, color='red', width=0.4, label='Women', alpha=0.3)

plt.ylabel("Rating")
plt.title("Movie Ratings by Gender")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(movie_type, men_rating, color='k', width=0.4, label='Men')
plt.bar(movie_type, women_rating, color='g', width=0.4, label='Women', bottom=men_rating)

plt.ylabel("Rating")
plt.title("Movie Ratings by Gender")
plt.legend()
plt.show()
```

# Box plotting

```python
# Box plotting
np.random.seed(10) 
n1=np.random.normal(100,10,200)
n2=np.random.normal(80,30,200)
n3=np.random.normal(70,50,200)
n4=np.random.normal(90,20,200)

data=[n1,n2,n3,n4]

# fig, ax = plt.subplots(figsize=(10, 5))
# ax.boxplot(data)
# plt.show()

plt.boxplot(data)
plt.show()
```

# Violin Plotting


```python
plt.violinplot(data)
plt.show()
```

# Histogram Plotting


```python
a = np.array([7, 23, 45, 12, 3, 67, 34, 5, 89, 78, 56, 50, 73, 28, 76, 87, 95, 18, 47])

plt.figure(figsize=(10, 5))
plt.hist(a, bins=5, color='blue', edgecolor='black')
plt.xticks([0, 25, 50, 75, 100])
plt.title('Histogram Plot')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

ax.hist(a, bins=5, color='blue')
ax.set_xticklabels([0, 25, 50, 75, 100, 125])
plt.show()
```

```python
# Hist Type
# 1) Bar
# 2) Step 
# 3) Stepfilled
# 4) Barstacked

plt.figure(figsize=(10, 5))
plt.hist(a, bins=5, color='blue', histtype='step')
plt.xticks([0, 25, 50, 75, 100])
plt.title('Histogram Plot')
plt.show()
```

# Scatter Shot Plotting

```python
boys_grade= [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
girls_grade= [82, 88, 80, 85, 90, 87, 93, 86, 89, 91]

range=[0,20,30,40,50,60,70,80,90,100]

plt.scatter(range,boys_grade, label='Boys', color='blue')
plt.scatter(range,girls_grade, label='Girls', color='red')
plt.legend()
plt.title('Grade Distribution')
plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.show()
```
